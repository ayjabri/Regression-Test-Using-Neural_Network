# Plot normal distribution in 3D space

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal

plt.style.use('seaborn-notebook')
get_ipython().run_line_magic('matplotlib', 'inline')
```

First we generate a set of 500 points (x,y) in range of -10, 10.  
Use numpy to generate a mesh of X,Y each will have 500x500 shape.  
Then create an empty position array where to stack X and Y in it

```python
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
X,Y = np.meshgrid(x,y)
pos = np.empty((500,500,2))
pos[:,:,0]=X;pos[:,:,1]=Y
```

### Choose a mean $\mu$ and variance $\sigma^2$:

```python
mean=[0,0];var = np.diag([4,12])
```

### Calculate the probablity of each point in the mesh

Use `multivariate_normal` function from `scipy` library to do that in one shot!

```python
Z = multivariate_normal.pdf(pos,mean=mean,cov=var)
```

### Plot the results using matplotlib 3d function:


```python
fig = plt.figure(num=1, figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X,Y,Z)
ax.contour(X,Y,Z, offset=0);
```
